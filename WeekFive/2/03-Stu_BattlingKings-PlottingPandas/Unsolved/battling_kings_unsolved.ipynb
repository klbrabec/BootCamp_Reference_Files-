```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
# Read CSV
kings_df = pd.read_csv("Resources/got.csv")
kings_df.head()
```

```python
# Get attacker and defender king data
attacker_king = kings_df["attacker_king"].value_counts()
defender_king = kings_df["defender_king"].value_counts()

attacker_king
```

```python
defender_king
```

```python
battle_df = pd.DataFrame({"Attacker": attacker_king, "Defender": defender_king}).fillna(0)
```

```python
# Get total battle data
battle_data = attacker_king.add(defender_king, fill_value=0)

battle_data
```

```python
# Configure plot and ticks
battle_data.plot(kind="bar",facecolor = 'green')

# Set textual properties
plt.title("The Bloodthirst of Kings")
plt.ylabel("Number of Battles Participated In")
plt.xlabel("King") #is there a way to line up the label with the tick?
plt.xticks(rotation=45)

# Resize plot to display labels
plt.tight_layout()
```
