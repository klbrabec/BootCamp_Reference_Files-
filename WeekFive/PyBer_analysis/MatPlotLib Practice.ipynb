```python
#import dependencies 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker ="*", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()

# set the X axis limit
# plt.xlim(0,5)

#set the graph size 
#plt.figure(figsize=(20,10))
```

```python
#skill drill - OO approach 
#example/formatting - https://matplotlib.org/matplotblog/posts/pyplot-vs-object-oriented-interface/
# formatting - https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html

fig, ax1 = plt.subplots()

ax1.plot (x_axis, y_axis,"green", marker = "D")
ax1.set_title("PyBer Fare by Month")
ax1.set_xlabel("Date")
ax1.set_ylabel("Fare($)")
ax1.legend("Boston")
ax1.grid()

fig.set_size_inches (7,5)
fig.set_dpi(100)

plt.show()


```

```python
#skill drill 
# Change the color of the markers to sky blue.
    #https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py
# Change the size of the markers to 5 times each data point.
# Make the color 20% transparent.
    #http://www.learningaboutelectronics.com/Articles/How-to-change-the-transparency-of-a-graph-plot-in-matplotlib-with-Python.php#:~:text=Matplotlib%20allows%20you%20to%20adjust,plot%20using%20the%20alpha%20attribute.&text=If%20you%20want%20to%20make,make%20alpha%20greater%20than%201.
# Add a black edge color to the circles.
    #https://vedexcel.com/plot-bubble-chart-in-python/#Bubble_Chart_output
# Make the linewidth of the circles 2 points.
    #https://www.python-graph-gallery.com/271-custom-your-bubble-plot
# Add a legend for the city of Boston.
# Add a title.
# Switch the axis so that the Fare($) data is on the x-axis.
# Add a limit for the x-axis data.
# Invert the y-axis so that January is at the top.

fig, ax = plt.subplots()

ax.scatter(y_axis, x_axis, c = "skyblue", alpha=.80, edgecolors='k',  linewidth=2,  s=[i * 5 for i in y_axis])
ax.legend("Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0, 50)
ax.invert_yaxis()

```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(y_axis,
       explode = explode_values, 
       labels=x_axis)
plt.show()
```

```python
# Add a percentage to one decimal place to each wedge of the pie.
# Increase the figure size to 8x8.
    #https://stackoverflow.com/questions/49199164/increasing-pie-chart-size-with-matplotlib-radius-parameter-appears-to-do-nothin
# Explode the two highest percentage months.
# Add a shadow.
    #https://stackoverflow.com/questions/41642126/matplotlib-customize-pie-chart-labels-and-dropshadow
# Add a start angle so that January is at the top.
    #https://stackoverflow.com/questions/41642126/matplotlib-customize-pie-chart-labels-and-dropshadow
# Reverse the order so that the month order is in a clockwise direction.
    #https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.pie.html
# Add new colors of your own choosing or use the colors from the previous pie chart.
   #https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.pie.html

explode_values = (0, 0, 0.2, 0, 0, 0, 0.3, 0.0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

fig, ax = plt.subplots(figsize = (8,8))
ax.pie(y_axis,
       explode = explode_values, 
       labels=x_axis, 
       colors = colors,
       autopct='%.1f%%',
       shadow=True, 
       startangle = 90, 
       counterclock = False
      )
plt.show()
```

```python
#import dependencies 

import matplotlib.pyplot as plt 
import statistics 
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=5)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=5)
plt.show()
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
