```python
import_file3 = os.path.join("avg_rain_state.csv")
rain_data = pd.read_csv(import_file3)
df = pd.DataFrame(rain_data)

#create graph 
tick_locations = [value for value in df["State"]]
                  
plt.figure(figsize=(20,3))
                  
plt.bar(df["State"],df["Inches"])
plt.xticks(tick_locations, df["State"], rotation='vertical')
           
plt.title("Rain in Inches by State")
plt.xlabel("State")
plt.ylabel("Rain")
plt.show() 
```
